## 5 - Mandelbrot

```python
def mandelbrotC(c, n, z=0):

    if n == 0:
        return z

    z2 = z**2 + c

    return mandelbrotC(c, n - 1, z2)
```

```python
mandelbrotC(complex(0, 1.1), 5)
```

```python
def value_in_iteration10(c):
    return mandelbrotC(c)
```
